```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import pandas
import random
import seaborn

matplotlib.rcParams['axes.labelsize'] = 16
matplotlib.rcParams['xtick.labelsize'] = 12
matplotlib.rcParams['ytick.labelsize'] = 12
seaborn.set_context({"figure.figsize": (12, 8)})
#specieslist = ['Dmel', 'Dpul', 'Agam', 'Pdom']
#speciesnames = ['Drosophila melanogaster', 'Daphnia pulex', 'Anopheles gambiae', 'Polistes dominula']
specieslist = ['Scer', 'Cele', 'Crei', 'Mtru', 'Dmel', 'Agam', 'Dpul', 'Xtro', 'Drer', 'Mmus', 'Hsap']
speciesnames = ['Saccharomyces cerevisiae', 'Caenorhabditis elegans', 'Chlamydomonas reinhardtii',
                'Medicago truncatula', 'Drosophila melanogaster', 'Anopheles gambiae', 'Daphnia pulex',
                'Xenopus tropicalis', 'Danio rerio', 'Mus musculus', 'Homo sapiens']
#specieslist = ['Scer', 'Cele', 'Crei', 'Vcar', 'Mtru', 'Grai', 'Dmel', 'Agam', 'Pdom', 'Dpul', 'Xtro', 'Drer', 'Mmus', 'Hsap']
#speciesnames = ['Saccharomyces cerevisiae', 'Caenorhabditis elegans', 'Chlamydomonas reinhardtii', 'Volvox carteri', 'Medicago truncatula', 'Gossypium raimondii', 'Drosophila melanogaster', 'Anopheles gambiae', 'Polistes dominula', 'Daphnia pulex', 'Xenopus tropicalis', 'Danio rerio', 'Mus musculus', 'Homo sapiens']
data = pandas.read_table('compactness.tsv')
palette = seaborn.color_palette("hls", len(specieslist))
#random.seed(54321)
#random.shuffle(palette)
```

```python
plots = list()
for i, species in enumerate(specieslist):
    #if species not in ['Dpul']: # ['Dpul', 'Grai']:
    #    continue
    specdata = data.loc[(data.Species == species)]
    plot = plt.scatter(specdata['AvgPercOcc'], specdata['AvgPercGene'], color=palette[i])
    plots.append(plot)
_ = plt.xlim(0, 1)
_ = plt.ylim(0, 1)
_ = plt.xlabel('Fraction of Sequence Occupied by miLoci')
_ = plt.ylabel('Fraction of Genes Contained in miLoci')
_ = plt.legend(plots, speciesnames, scatterpoints=1,
           loc='lower right',
           ncol=2,
           fontsize=12)
_ = plt.show()
```
